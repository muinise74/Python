### 7. 두 정수를 입력 받아 최대 공약수와 최소 공배수 구하기

```python
print('GCD와 LCM 구하기')
num1 = int(input("첫 번째 정수 입력 : "))
num2 = int(input("두 번째 정수 입력 : "))
temp1,temp2 = num1,num2
gcd,lcm = num1,num2
while(True):
    if(num1 > num2):
        num1 -= num2
    elif(num2 > num1):
        num2 -= num1
    else:
        gcd = num1
        break
lcm = int(temp1*temp2/gcd)
print('GCD :',gcd)
print('LCM :',lcm)
```

### 2. num1,num2,op를 매개변수로 받아 num1과 num2를 op에 맞게 연산한 결과값을 반한해주는 cal함수 정의 하기

```python
def cal(num1,num2,op):
    if (type(num1)==type(num2)==int):
        if (op == '+'):
            return num1+num2
        elif (op == '-'):
            return num1-num2
        elif (op == '*'):
            return num1*num2
        elif (op == '/'):
            return num1/num2
        elif (op == '//'):
            return num1//num2
        elif (op == '%'):
            return num1%num2
        elif (op == '**'):
            return num1**num2
        else:
            return '정의되지 않은 연산자 입니다.'
    else:
        return '연산을 수행할 수 없습니다.'
```

```python
num1,num2,num3,op1,op2 = 1,2,'3','*','4'
print(cal(num1,num2,op1))
print(cal(num1,num3,op1))
print(cal(num1,num2,op2))
```

### 3. 문자열을 입력받아 알파벳 순서대로 정렬하여 반환하는 함수를 정의하기 단, 대소문자도 구분

```python
def reverseStr(s):
    list1 = list(s)
    #sort함수를 사용하면 되지만 그런 문제가 아니므로 사용하지 않겠습니다.   
    #list 변수명으로 쓰면 지금처럼 list함수 쓸때에는 컴퓨터가 헷갈려서 오류 날수 있으니까 쓰지 맙시다.
    list2 = []
    for i in list1:
        list2.append(ord(i))
    for i in range(len(list2)):
        min = list2[i]
        index = i
        for j in range(i+1,len(list2)):
            if min > list2[j]:
                min = list2[j]
                index = j
        list2[index] = list2[i]
        list2[i] = min
    result = ''
    for i in list2:
        result += chr(i)
    return result
```
