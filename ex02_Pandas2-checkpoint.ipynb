```python
import pandas as pd
```

```python
df2015 = pd.read_csv('2015.csv',encoding = 'EUC-KR',index_col ='관서명')
df2015
```

```python
df2016 = pd.read_csv('2016.csv',encoding = 'EUC-KR',index_col ='관서명')
df2016
```

```python
df2017 = pd.read_csv('2017.csv',encoding = 'EUC-KR',index_col ='관서명')
df2017
```

```python
df2015.loc[:,'2015총계'] = df2015.loc[:,'살인':'폭력'].sum(axis = 1)
df2015
```

```python
df2016.loc[:,'2016총계'] = df2016.loc[:,'살인':'폭력'].sum(axis = 1)
df2016
```

```python
df2017.loc[:,'2017총계'] = df2017.loc[:,'살인':'폭력'].sum(axis = 1)
df2017
```

```python
# drop
# inplace 속성
# 삭제하고 원래 데이터에 그 결과를 반영(True)해주는 속성
df2017 = df2017.drop('광주지방경찰청')
df2017
```

```python
# 컬럼삭제
df2015.drop('구분',axis = 1)
# RollBack 불가 다시 쓰고 싶으면 처음부터 다시 해야 한다...
```

```python
df2015_result = pd.DataFrame(df2015.loc[df2015.loc[:,'구분']=='발생건수','2015총계'])
df2015_result
```

```python
df2016_result = pd.DataFrame(df2016.loc[df2016.loc[:,'구분']=='발생건수','2016총계'])
df2016_result
```

```python
df2017_result = pd.DataFrame(df2017.loc[df2017.loc[:,'구분']=='발생건수','2017총계'])
df2017_result
```

```python
df2015_result.rename(columns = {'2015총계':'총계'},inplace = True)
df2015_result
```

```python
df2016_result.rename(columns = {'2016총계':'총계'},inplace = True)
df2016_result
```

```python
df2017_result.rename(columns = {'2017총계':'총계'},inplace = True)
df2017_result
```

```python
df_inc_1516= (df2016_result-df2015_result)/df2015_result*100
df_inc_1516
```

```python
df_inc_1617= (df2017_result-df2016_result)/df2016_result*100
df_inc_1617
```

```python
df_result = pd.concat([df2015_result,df_inc_1516,df2016_result,df_inc_1617,df2017_result],axis = 1)
df_result.columns = ['2015 총계','15-16 증가율','2016 총계','16-17 증가율','2017 총계']
df_result
```
