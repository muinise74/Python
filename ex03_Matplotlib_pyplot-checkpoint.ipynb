```python
import matplotlib.pyplot as plt
import math
```

```python
# plot 함수: 일적선으로 되어있는 Lineplot 그리기
# plt.plot(x축,y축)
y = [2,4,6]
# Graph 그리기
plt.plot(y)
# Graph 출력
plt.show()
```

```python
# x축 설정하여 Graph 그리기
x,y = [1,2,3],[2,4,6]
plt.plot(x,y, ls = '--')
plt.show()
```

```python
# x축 설정하여 Graph 그리기
x,y = [2,4,6,8],[4,6,8,3]
y2 = [5,6,3,9]
plt.plot(x,y)
plt.plot(x,y2, ls = ":")
plt.show()
```

```python
plt.plot(x,y, marker = 'o')
plt.show()
```

```python
plt.plot(x,y, lw = math.pi)
plt.show()
```

```python
plt.plot(x,y, color = '#ffffff', marker = 'o', mfc = '#000000', mec = '#888888')
plt.show()
```

```python
x=[0,1,2,3,4,5,6]
y=[1,4,5,8,9,5,3]
plt.plot(x,y)
plt.show()
```

```python
x=[0,1,2,3,4,5,6]
y=[1,4,5,8,9,5,3]
plt.plot(x,y, marker = 'o', ls = '--', color = 'green')
plt.show()
```

```python
x=[0,1,2,3,4,5,6]
y=[1,4,5,8,9,5,3]
plt.plot(x,y, marker = 'D', ls = ':',color = 'red',mfc = 'blue',ms = 10)
plt.show()
```

```python
y = [1,4,5,8,9,5,3]
plt.plot(y, marker='o', ls = '--', color = 'green', mfc = 'red', ms = '10', mew = '3', lw = '5')
plt.show()
```
